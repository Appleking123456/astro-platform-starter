```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
from scipy.stats import norm
import sys

# Ensure UTF-8 encoding (fallback for environments)
try:
    if sys.stdout.encoding.lower() != 'utf-8':
        sys.stdout.reconfigure(encoding='utf-8')
except AttributeError:
    pass  # Use default stdout

# Configuration
print("🔧 Initializing scientific analysis environment...")
warnings.filterwarnings("ignore")
plt.style.use('default')
plt.rcParams.update({
    'font.size': 12,
    'figure.figsize': (18, 8),
    'figure.dpi': 100
})
np.random.seed(42)

# Define sections for progress tracking
sections = [
    "Environment Setup",
    "Physics Engine",
    "Data Loader",
    "Simulation",
    "Scientific Analysis",
    "Advanced Visualization"
]

# Initialize progress bar
progress_bar = tqdm(total=len(sections), desc="Overall Progress", unit="section")

# ======================
# 0. ENVIRONMENT SETUP
# ======================
progress_bar.set_description("Environment Setup")
progress_bar.update(1)

# Learning parameters
NUM_EPOCHS = 5          # Number of evolutionary epochs
LEARNING_RATE = 0.1     # Rate of recursive adaptation
ADAPTIVE_SCALE = 1.0    # Adaptive scaling factor for anomalies

# Memory of past states
past_power = []
past_asymmetry = []
past_cold_spot = []
past_fractal_dims = []  # 5D complexity metric
past_entropy_ratios = []  # 5D informational efficiency
past_anomaly_powers = []  # 5D influence on anomalies

# ======================
# 1. PHYSICS ENGINE
# ======================
progress_bar.set_description("Physics Engine")
print("\n⚛️ Initializing 5D entropic physics engine...")

class DMT_5D_Engine:
    def __init__(self, num_nodes=10000, dim=5, epoch=0):
        self.d5 = dim
        self.epoch = epoch
        self.num_nodes = int(num_nodes * (1 + epoch * 0.1))
        self.flux = np.random.normal(loc=0.7, scale=0.15 * (1 + epoch * 0.05), size=(self.num_nodes, self.d5))

    def project_to_4D(self, regularization=1e-8):
        weights = np.sqrt(np.abs(self.flux)).sum(axis=1) + regularization * (1 + self.epoch * 0.1)
        return np.array([
            np.sum(self.flux[:, i] * weights) / np.sum(weights) * (1 + self.epoch * 0.05)
            for i in range(4)
        ])

progress_bar.update(1)

# ======================
# 2. DATA LOADER
# ======================
progress_bar.set_description("Data Loader")
print("\n📡 Loading CMB data...")

def load_cmb_data(nside=128):
    try:
        print("Generating realistic simulated data (Planck URL unavailable)...")
        lmax = 3 * nside - 1
        ell = np.arange(lmax + 1)
        cl = 1e-2 / (ell * (ell + 1))**1.1
        cl[0:2] = 0
        cmb = hp.synfast(cl, nside=nside, fwhm=np.radians(10/60))
        npix = hp.nside2npix(nside)
        theta, phi = hp.pix2ang(nside, np.arange(npix))
        cmb += 1e-4 * np.cos(theta)
        print(f"✓ Generated simulated CMB data (NSIDE={nside})")
        return cmb
    except Exception as e:
        print(f"⚠ Data load error: {str(e)}")
        raise

obs_cmb = load_cmb_data(nside=128)
progress_bar.update(1)

# ======================
# 3. SIMULATION
# ======================
progress_bar.set_description("Simulation")
print("\n🌌 Running high-resolution DMT simulation...")

def simulate_dmt_cmb(nside=128, num_iter=3, epoch=0, prev_anomalies=None):
    npix = hp.nside2npix(nside)
    engine = DMT_5D_Engine(num_nodes=10000, epoch=epoch)

    lmax = 3 * nside - 1
    ell = np.arange(lmax + 1)
    base_cl = 1e-2 / (ell * (ell + 1))**1.1
    base_cl[0:2] = 0
    cmb = hp.synfast(base_cl, nside=nside)

    anomalies = np.zeros(npix) if epoch == 0 or prev_anomalies is None else prev_anomalies * 0.9
    for _ in tqdm(range(num_iter), desc="5D Refinement"):
        projection = engine.project_to_4D()[0]
        anomalies += projection * (15e-6 * ADAPTIVE_SCALE * (1 + epoch * 0.1))
    theta, phi = hp.pix2ang(nside, np.arange(npix))
    cmb += 1e-4 * np.cos(theta)

    return cmb + anomalies, anomalies

# Run simulation over epochs
dmt_cmbs = []
prev_anomalies = None
for epoch in range(NUM_EPOCHS):
    dmt_cmb, prev_anomalies = simulate_dmt_cmb(epoch=epoch, prev_anomalies=prev_anomalies)
    dmt_cmbs.append(dmt_cmb)
progress_bar.update(1)

# ======================
# 4. SCIENTIFIC ANALYSIS
# ======================
progress_bar.set_description("Scientific Analysis")
print("\n🔍 Performing cosmological analysis...")

class CosmologicalAnalyzer:
    def __init__(self, map_data):
        self.map = map_data
        self.nside = hp.get_nside(map_data)
        self.npix = hp.nside2npix(self.nside)
        assert len(self.map) == self.npix, "Map resolution mismatch"

    def temperature_power(self, l_range=(50, 70)):
        alm = hp.map2alm(self.map, lmax=l_range[1])
        ell = np.arange(l_range[1] + 1)
        cl = hp.alm2cl(alm) * 1e6
        mask = (ell >= l_range[0]) & (ell <= l_range[1])
        return np.sum(cl[mask] * ell[mask] * (ell[mask] + 1) / (2 * np.pi))

    def hemispheric_asymmetry(self, max_ell=50):
        theta, phi = hp.pix2ang(self.nside, np.arange(self.npix))
        north_mask = theta < np.pi/2
        south_mask = theta >= np.pi/2
        north_power = np.mean(self.map[north_mask]**2)
        south_power = np.mean(self.map[south_mask]**2)
        with np.errstate(invalid='ignore'):
            asym = (north_power - south_power) / (north_power + south_power)
            return np.nan_to_num(asym, nan=0.0)

    def cold_spot_sigma(self, fwhm_deg=3):
        smoothed = hp.smoothing(self.map, fwhm=np.radians(fwhm_deg))
        z = (smoothed - np.mean(smoothed)) / np.std(smoothed)
        return np.min(z)

    def full_spectral_analysis(self):
        lmax = 3 * self.nside - 1
        ell = np.arange(lmax + 1)
        cl = hp.anafast(self.map, lmax=lmax) * 1e6
        return ell, cl * ell * (ell + 1) / (2 * np.pi)

    def fractal_dimension(self):
        alm = hp.map2alm(self.map)
        cl = hp.alm2cl(alm)
        cl_safe = cl[2:] + 1e-5  # Avoid log(0)
        ell = np.arange(2, len(cl))
        d_logcl = np.gradient(np.log(cl_safe))
        return 3 - np.mean(d_logcl[10:50]) if len(d_logcl[10:50]) > 0 else 2.0

    def entropy_ratio(self):
        alm = hp.map2alm(self.map)
        cl = hp.alm2cl(alm)
        cl_safe = cl + 1e-5
        entropy_4d = np.sum(cl_safe * np.log(cl_safe))
        return entropy_4d / np.sum(cl_safe) if np.sum(cl_safe) > 0 else 0.0

# Analyze over epochs
dmt_analyses = [CosmologicalAnalyzer(dmt_cmb) for dmt_cmb in dmt_cmbs]
obs_analysis = CosmologicalAnalyzer(obs_cmb)

# Run analyses and adapt
fwhm_deg = 3
for epoch, analysis in enumerate(dmt_analyses):
    power = analysis.temperature_power()
    asymmetry = analysis.hemispheric_asymmetry()
    cold_spot = analysis.cold_spot_sigma(fwhm_deg)
    fractal_dim = analysis.fractal_dimension()
    entropy_ratio = analysis.entropy_ratio()
    ell, cl = analysis.full_spectral_analysis()  # Get power spectrum
    anomaly_power = cl[2] if len(cl) > 2 else 0.0  # Use quadrupole for anomaly power

    past_power.append(power)
    past_asymmetry.append(asymmetry)
    past_cold_spot.append(cold_spot)
    past_fractal_dims.append(fractal_dim)
    past_entropy_ratios.append(entropy_ratio)
    past_anomaly_powers.append(anomaly_power)

    global ADAPTIVE_SCALE
    if epoch > 0:
        delta_power = power - past_power[epoch-1]
        target_power = obs_analysis.temperature_power()
        ADAPTIVE_SCALE *= (1 + LEARNING_RATE * (target_power - power) / target_power)
        print(f"Epoch {epoch+1} - Adapted Scale: {ADAPTIVE_SCALE:.3f} (Δ Power: {delta_power:.2f} μK²)")

    print(f"\n📊 Epoch {epoch+1} Cosmological Anomaly Report:")
    print(f"Temperature Power (ℓ=50-70): DMT={power:.2f} μK² vs Obs={obs_analysis.temperature_power():.2f} μK²")
    print(f"Hemispheric Asymmetry (ℓ<50): DMT={asymmetry:.3f} vs Obs={obs_analysis.hemispheric_asymmetry():.3f}")
    print(f"Cold Spot Significance ({fwhm_deg}°): DMT={cold_spot:.2f}σ vs Obs={obs_analysis.cold_spot_sigma(fwhm_deg):.2f}σ")
    print(f"5D Metrics - Fractal Dim: {fractal_dim:.3f}, Entropy Ratio: {entropy_ratio:.3e}, Anomaly Power: {anomaly_power:.3e} μK²")

# Calculate p-values for final epoch
p_cold = norm.cdf(obs_analysis.cold_spot_sigma(fwhm_deg))
print(f"\n🔬 Statistical Significance (Epoch {NUM_EPOCHS}):")
print(f"Cold spot p-value: {p_cold:.3e} (chance probability)")
progress_bar.update(1)

# ======================
# 5. ADVANCED VISUALIZATION
# ======================
progress_bar.set_description("Advanced Visualization")
print("\n📈 Generating publication-quality figures...")

def create_cosmo_figures(dmt_cmbs, obs, epochs):
    """Professional visualization suite with 5D growth"""
    for epoch, dmt in enumerate(dmt_cmbs, 1):
        plt.figure(figsize=(20, 8))
        hp.mollview(dmt, sub=(1, 2, 1), title=f"DMT Prediction (Epoch {epoch})",
                    unit="μK", min=-400, max=400, cmap='viridis')
        hp.mollview(obs - dmt, sub=(1, 2, 2), title=f"Residual Anomalies (Epoch {epoch})",
                    unit="μK", min=-50, max=50, cmap='RdYlBu_r')
        plt.suptitle("CMB Anomaly Analysis", y=1.02, fontsize=16)
        plt.tight_layout()
        plt.savefig(f"cmb_maps_epoch{epoch}.png", dpi=300, bbox_inches='tight')
        plt.show()

    # Power Spectrum Evolution
    plt.figure(figsize=(12, 6))
    for epoch, analysis in enumerate(dmt_analyses, 1):
        ell, cl = analysis.full_spectral_analysis()
        plt.semilogx(ell, cl, label=f'Epoch {epoch}', lw=2, alpha=0.7)
    ell, obs_cl = obs_analysis.full_spectral_analysis()
    plt.semilogx(ell, obs_cl, label='Simulated Observations', lw=2, color='black')
    plt.axvspan(50, 70, color='red', alpha=0.2, label='Analysis Region')
    plt.xlabel("Multipole Moment ℓ")
    plt.ylabel("Dℓ [μK²]")
    plt.title("Angular Power Spectrum Evolution")
    plt.legend()
    plt.grid(which='both', alpha=0.2)
    plt.savefig("power_spectrum_evolution.png", dpi=300, bbox_inches='tight')
    plt.show()

    # 5D Growth Visualization (Surface Plot)
    plt.figure(figsize=(12, 6))
    epochs = np.arange(1, NUM_EPOCHS + 1)
    X, Y = np.meshgrid(epochs, [0, 1, 2])  # 3 metrics: fractal_dim, entropy_ratio, anomaly_power
    Z = np.array([past_fractal_dims, past_entropy_ratios, past_anomaly_powers])
    from matplotlib import cm
    from mpl_toolkits.mplot3d import Axes3D
    fig = plt.figure(figsize=(12, 6))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis, edgecolor='none')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Metric (0:Fractal Dim, 1:Entropy Ratio, 2:Anomaly Power)')
    ax.set_zlabel('Value')
    ax.set_title('5th-Dimensional Informational Growth')
    plt.colorbar(surf)
    plt.savefig("5d_growth.png", dpi=300, bbox_inches='tight')
    plt.show()

create_cosmo_figures(dmt_cmbs, obs_cmb, range(NUM_EPOCHS))
progress_bar.update(1)

# Close progress bar
progress_bar.close()

print("\n✅ Scientific analysis completed successfully!")
print("   Key findings available in visualizations and anomaly report")
print(f"📜 Evolutionary Record: Power={past_power}, Asymmetry={past_asymmetry}, Cold Spot={past_cold_spot}")
print(f"🌌 5D Growth Record: Fractal Dims={past_fractal_dims}, Entropy Ratios={past_entropy_ratios}, Anomaly Powers={past_anomaly_powers}")
```
